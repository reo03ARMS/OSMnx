```python
import pandas as pd

df = pd.read_csv('../demo_michiten_location.csv')
df.columns
```

```python
sub_df=df[["id","created_at","updated_at","time_target_lat","time_target_lng","time_target_time"]]
sub_df.head(5)
```

```python
#緯度
lat = sub_df.iloc[1]["time_target_lat"]
lat
```

```python
#経度
lng = sub_df.iloc[1]["time_target_lng"]
lng
```

```python
import osmnx as ox
import networkx as nx

G=nx.Graph()
#座標を指定し地図データをダウンロード
G = ox.graph_from_point(center_point=(lat,lng), network_type='drive', dist=1000)
```

```python
#グラフを可視化する
ox.plot_graph(G,node_color='#87cefa',save=True,filepath='All_request.jpg')
```

```python
nodes, edges = ox.graph_to_gdfs(G)
nodes.to_csv("node_road.csv")
edges.to_csv("edges_road.csv")

# nodes: pointが交差点　信号なし
# nodes: traffic_signals,point 信号あり交差点
```

```python
#地図の表示
import folium as fol

map1 = fol.Map(
    #初期位置の設定
    location=[lat,lng],
    
    #初期表示の拡大設定
    zoom_start=15,
    
    #地図のスタイル
    tiles= "OpenStreetMap"
)
map1
```
